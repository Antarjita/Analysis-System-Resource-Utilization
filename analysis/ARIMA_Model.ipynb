```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names=["Time","cpu_utilization","processor_time","user_time","nice_time","idle_time","memory_utilization","used_mem","total_mem","free_mem","available_mem","pagetable","vmused","num_pages","pf_before","pf_persec","ps_in_persec","ps_out_psec","pread_persec","pwrite_persec","p_persec"]
data=pd.read_csv("overallinfo-1.csv",names=column_names)
```

```python
data
```

```python
data.drop(['p_persec','ps_in_persec','ps_out_psec'],axis='columns',inplace=True)
```

```python
data
```

```python
df=data
df.drop(['user_time','nice_time','idle_time','processor_time','used_mem','total_mem','free_mem','available_mem','pagetable','vmused','num_pages','pf_before','pf_persec','pread_persec','pwrite_persec'],axis='columns',inplace=True)
```

```python
df
```

```python
df['Time']=pd.to_datetime(df['Time'])
df
```

```python
df=df.set_index('Time')
df.index = pd.DatetimeIndex(df.index).to_period('H')
df.index
```

```python
cpu=df[ 'cpu_utilization']
cpu.head()
```

```python
from statsmodels.tsa.stattools import adfuller
X = cpu.values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
from sklearn.model_selection import train_test_split

cputrain, cputest = train_test_split(cpu, test_size=0.2)
type(cputrain)
```

```python
cpu.plot(figsize=(15,6))
plt.show()
```

```python
from pylab import rcParams
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
rcParams['figure.figsize']=18,8
pd.plotting.autocorrelation_plot(cputrain)
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
# fit model
model = ARIMA(cputrain, order=(1,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
history=[x for x in cputrain]
predictions =list()
for t in range(len(cputest)):
	model = ARIMA(history, order=(1,1,0))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = cputest[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(cputest, predictions)
print('Test MSE: %.3f' % error)

```
