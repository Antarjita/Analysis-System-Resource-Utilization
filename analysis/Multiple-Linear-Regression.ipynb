```python
#Prediction of cpu utilization 
import pandas as pd
from sklearn import linear_model
df=pd.read_csv("overallinfo.csv")
```

```python
df
```

```python
column_names=["time","cpu_utilization","processor_time","user_time","nice_time","idle_time","memory_utilization","used_mem","total_mem","free_mem","available_mem","pagetable","vmused","num_pages","pf_before","pf_persec","ps_in_persec","ps_out_psec","pread_persec","pwrite_persec","p_persec"]
df=pd.read_csv("overallinfo.csv",names=column_names)
```

```python
df
```

```python
df.dtypes
```

```python
df=df.drop(["time","memory_utilization","used_mem","total_mem","free_mem","available_mem","pagetable","vmused","num_pages","pf_before","pf_persec","ps_in_persec","ps_out_psec","pread_persec","pwrite_persec","p_persec"],axis=1)
```

```python
df_copy=df.copy()
training_data=df_copy.sample(frac=0.75)
test_data=df_copy.drop(training_data.index)
```

```python
training_data
```

```python
X=training_data[["processor_time","user_time","nice_time","idle_time"]]
y=training_data["cpu_utilization"]
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

```python
test_data=test_data.drop(["cpu_utilization"],axis=1)
```

```python
prediction=reg.predict(test_data)
```
